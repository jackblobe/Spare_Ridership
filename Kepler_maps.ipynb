```python
import pandas as pd
```

```python
dallas_base = gpd.read_file('Census_Block_Group_2010_TX.geojson')
```

```python
dallas_clusters=pd.read_csv('/Users/jacklobe/lighthouse/Final/clustered_CSV/dallas_kmeans.csv')
```

```python
dallas_bus=pd.read_csv('dallas_transitmap.csv')
```

```python
from keplergl import KeplerGl 
map_1 = KeplerGl(height=500)
map_1
```

```python
map_1.add_data(data=dallas_base, name='basemap')
```

```python
map_1.add_data(data=dallas_clusters,name='clusters')
map_1.add_data(data=dallas_bus,name='routes')
```

```python
map_1.save_to_html(file_name='kepler_clustered_dallas.html')
```

```python
swtran_cluster=pd.read_csv('clustered_CSV/swtran_clusters.csv')
swtran_race=pd.read_csv('racial_swtran.csv')
```

```python
map_2 = KeplerGl(height=500)
map_2.add_data(data=swtran_cluster, name='sw_clusters')
map_2.add_data(data=swtran_race,name='sw_race')
```

```python
map_2.save_to_html(file_name='kepler_clustered_swtran.html')
```
